```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the stock data into a pandas dataframe
df = pd.read_csv('AAPL.csv')
```

```python
# Define a function to identify the Hammer pattern
def is_hammer(row):
    # Check that the pattern exists
    if len(df) < 1:
        return False
    # Check that the pattern is a Hammer
    body = row['Close'] - row['Open']
    upper_shadow = row['High'] - row['Close']
    lower_shadow = row['Open'] - row['Low']
    if (body > 0) and (upper_shadow < 0.1*body) and (lower_shadow > 2*body):
        return True
    else:
        return False

# Identify the Hammer patterns in the data
hammer_mask = df.apply(is_hammer, axis=1)
hammer_dates = df[hammer_mask]['Date']
```

```python
# Print out the dates of the patterns
print('Dates of Hammer patterns:')
for date in hammer_dates:
    print(date)
```

```python
# Plot the stock data with the patterns highlighted
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(df['Date'], df['Close'], label='Closing Price')
ax.scatter(hammer_dates, df[hammer_mask]['Close'], color='g', label='Hammer Pattern')
ax.set_title('Stock data with Hammer patterns')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.legend()
plt.show()
```
